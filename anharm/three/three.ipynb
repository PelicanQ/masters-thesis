```python
from anharm.Hamiltonian import Hamil
import sympy as sp
H = Hamil(3, 5, "line")
keep = False
s = H.zzexpr("111", keep_second_coupling=keep)
# s = sp.collect(s, [sp.Symbol("g_{0,1}"),sp.Symbol("g_{1,2}")])
# s = s.subs(sp.Symbol(r"\Delta_{1,2}"), -sp.Symbol(r"\Delta_{0,1}"))
# s = s.subs(sp.Symbol(r"g_{1,2}"), sp.Symbol(r"g_{0,1}"))
# s = s.subs(sp.Symbol(r"\Delta_{0,2}"), 0)
s = s.subs(sp.Symbol(r"\alpha_{2}"), sp.Symbol(r"\alpha_{0}"))
s = s.subs(sp.Symbol(r"\alpha_{1}"), sp.Symbol(r"\alpha_{0}"))
s = H.split_deltas(s)
f,vars  = H.lambdify_expr(s) # alpha0 g01 g12 delta01 delta12
d01 = np.arange(-8,8,0.03)
d12 = np.arange(-8,8,0.03)
X,Y = np.meshgrid(d01,d12)
F = f(-1, 1, 1, X,Y)
F
# sp.simplify(s)
# sp.collect(s, [sp.Symbol(r"g_{0,1}")*sp.Symbol(r"g_{1,2}")])

plt.pcolor(d01,d12,F, norm=colors.SymLogNorm(linthresh=1e-6, vmin=-1e2, vmax=1e2, clip=False))
plt.colorbar()
plt.show()
```
