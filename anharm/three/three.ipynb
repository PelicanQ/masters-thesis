```python
from anharm.Hamiltonian import Hamil
import sympy as sp
from matplotlib import pyplot as plt, colors
import numpy as np
```

```python
H = Hamil(3, 4, "triang")
e = H.zzexpr("111", 3)-(H.zzexpr("101",3)+H.zzexpr("011",3)+H.zzexpr("110",3))
e = H.split_deltas(e)
f,vars = H.lambdify_expr(e)
#\alpha_{0}, \alpha_{1}, \alpha_{2}, g_{0,1}, g_{1,2}, g_{0,2}, \Delta_{0,1}, \Delta_{1,2}]
print(vars)
display(e)
xx = np.linspace(-10,10,200)
X,Y=np.meshgrid(xx,xx)
Z = f(-1,-1,-1, 0.4,0.4,0.4,X,Y)

@np.vectorize
def snapto0(v):
    if np.abs(v) < 1e-5:
        return 0.0
    return v
Z = snapto0(Z)
plt.pcolormesh(X,Y,Z, norm=colors.SymLogNorm(1e-1, vmin=-1e-1,vmax=1e-1))
plt.colorbar()
plt.show()
# what if the sum of zz always get's the sign of zzz right?
```
